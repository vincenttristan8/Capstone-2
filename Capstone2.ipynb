melakukan describe terhadap price untuk menentukan range harga yang paling prime untuk meningkatkan minat market. Melakukan boxplot agar memperjelas outliers yang muncul pada data. Dan pada analisa ini, menggunakan kondisi dimana minimum night = 1 agar data price yang diuji valid dan tidak terganggu oleh faktor lain nya. Dapat dilihat terdapat beberapa listingan yang menyebabkan outlier muncul. 

```python
filter = df[df['minimum_nights'] == 1]
lower_quantile = 0.05
upper_quantile = 0.95

pricequantiles = filter.groupby('room_type')['price'].quantile([lower_quantile, upper_quantile]).unstack()
filtered_data = filter.merge(pricequantiles, on='room_type')
filtered_data = filtered_data[
    (filtered_data['price'] >= filtered_data[lower_quantile]) &
    (filtered_data['price'] <= filtered_data[upper_quantile])]

descriptive = filtered_data.groupby('room_type')['price'].describe().round(1).T
print(descriptive)
```

melakukan cleaning terhadap outlier atau listingan dengan harga upnormal. Didapatkan harga optimal pada kisaran harga optimal untuk tipe akomodasi 'Seluruh rumah/apartemen' ditetapkan antara 1200 hingga 3069 Baht. Untuk 'Kamar hotel', kisaran yang disarankan sekitar 1000 hingga 3000 Baht, sedangkan 'Kamar pribadi' sebaiknya dihargai antara 900 hingga 2229 Baht. 'Kamar bersama' idealnya dihargai dalam kisaran 400 hingga 600 Baht. penentuan harga menggunakan data 25% dan 75% dengan maksud agar harga tidak underprice maupun overprice.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt

top_100_listings = df.nlargest(100, 'number_of_reviews')

all_listing_names = ' '.join(top_100_listings['name'].dropna())
tokens = word_tokenize(all_listing_names)
stop_words = set(stopwords.words('english'))
filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
top_words = Counter(filtered_tokens).most_common(20)
top_words_df = pd.DataFrame(top_words, columns=['Word', 'Count'])

plt.figure(figsize=(12, 8))
plt.bar(top_words_df['Word'], top_words_df['Count'])
plt.title('Top 20 Used Words in the "name" Column (Top 100 Listings by Number of Reviews)')
plt.xlabel('Word')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

menganalisa top penggunaan kata pada data dimana top 100 number_of_reviews dengan maksud top 100 listing paling diminati. didapatkan 20 kata-kata yang paling sering digunakan untuk listingan. sebagai masukkan bagi audience untuk melakukan listingan. Serta pada chart ini dapat dilihat berbagai fasilitas kamar yang dapat menjadi nilai tambah terhadap minat customer seperti dekat dengan MRT, memiliki pemandangan yang bagus, memiliki studio, adanya fasilitas wifi.
